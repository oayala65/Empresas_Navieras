```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Ruta del archivo
file_path = "TAGUATO_09_04_2025-13_17_56_keras.xlsx"

# Cargar el archivo y ver las hojas disponibles
xls = pd.ExcelFile(file_path)
xls.sheet_names
```

```python
# Cargar la hoja en un DataFrame
df = pd.read_excel(xls, sheet_name="Work",parse_dates=["DATE/TIME(UTC)"])

# Mostrar las primeras filas del DataFrame
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1)
```

```python
df.T
```

```python
df.columns
```

```python
# Graficar con Seaborn
plt.figure(figsize=(15, 5))
sns.lineplot(x=df["DATE/TIME(UTC)"], y=df["VELOCIDAD"], marker="o")
# Personalizar la grÃ¡fica
plt.xlabel("Fecha y Hora (UTC)")
plt.ylabel("Velocidad")
plt.title("Velocidad vs. Tiempo (06:30 - 06:40 UTC)")
plt.xticks(rotation=45)  # Rotar etiquetas del eje X para mejor visualizaciÃ³n
plt.grid(True)

# Mostrar la grÃ¡fica
plt.show()
```

```python
# Filtrar datos entre las 06:30 y 06:40 UTC
hora_inicio = pd.to_datetime("2025-03-25 06:30:00")
hora_fin = pd.to_datetime("2025-03-25 06:40:00")
df_filtrado = df[(df["DATE/TIME(UTC)"] > hora_inicio) & (df["DATE/TIME(UTC)"] < hora_fin)]

# Graficar con Seaborn
plt.figure(figsize=(15, 5))
sns.lineplot(x=df_filtrado["DATE/TIME(UTC)"], y=df_filtrado["VELOCIDAD"], marker="o")
# Personalizar la grÃ¡fica
plt.xlabel("Fecha y Hora (UTC)")
plt.ylabel("Velocidad")
plt.title("Velocidad vs. Tiempo (06:30 - 06:40 UTC)")
plt.xticks(rotation=45)  # Rotar etiquetas del eje X para mejor visualizaciÃ³n
plt.grid(True)

# Mostrar la grÃ¡fica
plt.show()
```
