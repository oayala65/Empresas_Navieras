```python
import pandas as pd

# Cargar el archivo Excel
file_path = "TAGUATO_19_03_2025-13_07_18_completo.xlsx"  # Reemplaza con la ubicación de tu archivo
df = pd.read_excel(file_path)

# Ver las primeras filas
df.head()
```

```python
# df.to_csv('taguato_completo.csv', index=False)
```

```python
df=df.drop('DATE/TIME(America/Asuncion)',axis=1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from gplearn.genetic import SymbolicRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Definir las variables de entrada y salida
X = df.drop(columns=["Flow Rate Prop. BR", "Flow Rate Prop. STB"])  # Usamos todas las variables excepto la salida
y_BR = df["Flow Rate Prop. BR"]
y_STB = df["Flow Rate Prop. STB"]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)


df_filled = df.fillna(df.median())

X = df_filled.drop(columns=["Flow Rate Prop. BR", "Flow Rate Prop. STB"])
y_BR = df_filled["Flow Rate Prop. BR"]
y_STB = df_filled["Flow Rate Prop. STB"]

# Dividir en entrenamiento y prueba
X_train, X_test, y_train_BR, y_test_BR = train_test_split(X, y_BR, test_size=0.2, random_state=42)
X_train, X_test, y_train_STB, y_test_STB = train_test_split(X, y_STB, test_size=0.2, random_state=42)

# Crear el modelo de regresión simbólica
model_BR = SymbolicRegressor(
    population_size=1000,   # Tamaño de la población en cada generación
    generations=100,        # Número de generaciones
    function_set=('add', 'sub', 'mul', 'div', 'sin', 'cos', 'sqrt', 'log'),  # Operaciones permitidas
    metric='mse',          # Minimizar el error cuadrático medio
    verbose=1,             # Mostrar progreso
    random_state=42
)

model_STB = model_BR


# Entrenar el modelo
model_BR.fit(X_train, y_train_BR)
model_STB.fit(X_train, y_train_STB)

# Hacer predicciones
y_pred_BR = model_BR.predict(X_test)
y_pred_STB = model_STB.predict(X_test)

# Evaluar el modelo
mse_BR = mean_squared_error(y_test_BR, y_pred_BR)
r2_BR = r2_score(y_test_BR, y_pred_BR)

mse_STB = mean_squared_error(y_test_STB, y_pred_STB)
r2_STB = r2_score(y_test_STB, y_pred_STB)

print(f"Error Cuadrático Medio (MSE): {mse_BR:.4f}")
print(f"Coeficiente de Determinación (R²): {r2_BR:.4f}")

print(f"Error Cuadrático Medio (MSE): {mse_STB:.4f}")
print(f"Coeficiente de Determinación (R²): {r2_STB:.4f}")


# Mostrar la ecuación encontrada
print("Ecuación simbólica encontrada:")
print(model_BR._program)

print("Ecuación simbólica encontrada:")
print(model_STB._program)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predicciones en el conjunto de prueba
y_pred_BR_test = model_BR.predict(X_test)
y_pred_STB_test = model_STB.predict(X_test)

# Cálculo de métricas para Flow Rate Prop. BR
mae_BR = mean_absolute_error(y_test_BR, y_pred_BR_test)
mse_BR = mean_squared_error(y_test_BR, y_pred_BR_test)
rmse_BR = np.sqrt(mse_BR)
r2_BR = r2_score(y_test_BR, y_pred_BR_test)

# Cálculo de métricas para Flow Rate Prop. STB
mae_STB = mean_absolute_error(y_test_STB, y_pred_STB_test)
mse_STB = mean_squared_error(y_test_STB, y_pred_STB_test)
rmse_STB = np.sqrt(mse_STB)
r2_STB = r2_score(y_test_STB, y_pred_STB_test)

# Mostrar resultados
print("**Errores para Flow Rate Prop. BR:**")
print(f"MAE: {mae_BR:.2f}")
print(f"MSE: {mse_BR:.2f}")
print(f"RMSE: {rmse_BR:.2f}")
print(f"R²: {r2_BR:.4f}\n")

print("**Errores para Flow Rate Prop. STB:**")
print(f"MAE: {mae_STB:.2f}")
print(f"MSE: {mse_STB:.2f}")
print(f"RMSE: {rmse_STB:.2f}")
print(f"R²: {r2_STB:.4f}")
```

```python
import seaborn as sns

# Generar datos de entrada para la gráfica
X_median = np.median(X_train, axis=0)  # Valores fijos en la mediana
X_fixed_BR = np.tile(X_median, (100, 1))
X_fixed_STB = np.tile(X_median, (100, 1))

# Modificar las columnas correspondientes
rack_br_idx = list(df.columns).index("Rack Prop. BR") - 1
rack_stb_idx = list(df.columns).index("Rack Prop. STB") - 1

x_range_BR = np.linspace(df["Rack Prop. BR"].min(), df["Rack Prop. BR"].max(), 100)
X_fixed_BR[:, rack_br_idx] = x_range_BR
y_pred_BR = model_BR.predict(X_fixed_BR)

x_range_STB = np.linspace(df["Rack Prop. STB"].min(), df["Rack Prop. STB"].max(), 100)
X_fixed_STB[:, rack_stb_idx] = x_range_STB
y_pred_STB = model_STB.predict(X_fixed_STB)

# Crear gráficos
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico para la hélice de babor (BR)
sns.scatterplot(x=df["Rack Prop. BR"], y=df["Flow Rate Prop. BR"], ax=axes[0], color="blue", alpha=0.5)
axes[0].plot(x_range_BR, y_pred_BR, color="black", linewidth=2)
axes[0].set_title("Regresión Lineal: Rack Prop. BR vs Flow Rate Prop. BR")
axes[0].set_xlabel("Rack Prop. BR (Ángulo)")
axes[0].set_ylabel("Flow Rate Prop. BR (Caudal)")

# Gráfico para la hélice de estribor (STB)
sns.scatterplot(x=df["Rack Prop. STB"], y=df["Flow Rate Prop. STB"], ax=axes[1], color="red", alpha=0.5)
axes[1].plot(x_range_STB, y_pred_STB, color="black", linewidth=2)
axes[1].set_title("Regresión Lineal: Rack Prop. STB vs Flow Rate Prop. STB")
axes[1].set_xlabel("Rack Prop. STB (Ángulo)")
axes[1].set_ylabel("Flow Rate Prop. STB (Caudal)")

plt.tight_layout()
plt.show()
```
